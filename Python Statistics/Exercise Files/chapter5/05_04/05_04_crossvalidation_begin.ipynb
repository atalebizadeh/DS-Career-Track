Refactored plotting so that it works on arbitrary datasets.

```python
def sc(data):
    return 1e-6 * data.population, data.region.map({'Africa': 'skyblue', 'Europe': 'gold',
                                                   'America': 'palegreen', 'Asia': 'coral'})

def plotdata(data=gdata):
    size, colors = sc(data)
    data.plot.scatter('age5_surviving','babies_per_woman',
                      c=colors,s=size,linewidths=0.5,edgecolor='k',alpha=0.5)
                      
def plotfit(fit,data=gdata):
    plotdata(data)

    size, colors = sc(data)
    pp.scatter(data.age5_surviving,fit.predict(data),
               c=colors,s=30,linewidths=0.5,edgecolor='k',marker='D')
```

Model formulas from the last two videos

```python
f_groupmeans = 'babies_per_woman ~ -1 + region'
f_surviving = 'babies_per_woman ~ -1 + region + age5_surviving'
f_srp = ('babies_per_woman ~ -1 + region + age5_surviving'
         '+ age5_surviving:region - age5_surviving + population')
```

```python
# To divide up our data we first shuffle it with a Panda's dataframe method sample(). 
shuffled = gdata.sample(len(gdata))
```

```python
# Then we use Panda's locator iloc[] which goes by row number to divide the rows in two
training, testing = shuffled.iloc[:len(gdata)//2], shuffled.iloc[len(gdata)//2:]
```

```python
# We now fit the model to the training data
surviving = smf.ols(f_surviving, data=training).fit()
```

```python
plotfit(surviving, training)
```

```python
surviving.mse_resid
```

```python
# We now try out the model on the testing data
plotfit(surviving,testing)
```

### To obtain the mean square of error, we need to do a little work
1. Compute the residuals by applying modern prediction to the testing dataset. We subtract the observe data
2. We compute the number of degrees of freedom which is the length of the data minus the number of patameters in the model minus 1 since the number of parameters don't count the constant, but we should.
3. Finally, we square up the residuals, sum them, and divide by the number of degrees of freedom

```python
resid = surviving.predict(testing) - testing.babies_per_woman
df = len(testing) - surviving.df_model - 1
np.sum(resid**2)/ df
```

### Let's automate this procedure and use it on the three models.
#### Since the results depend on the shuffling and are rather noisy, we'll repeat a few times and take the average.
1. Let's make a function that's generic in terms of data, formula, response variable that we're looking at, and of the number of repitions
2. We repeat in times, and we are going to take the code from above and modify them properly.
3. We shuffle
4. We divide in two
5. We train
6. And we compute the mean square of error
7. We grab an arbitrary response viable column
8. And finally we collect all these numbers in a python list and return its average.

```python
def cross_validate(data, formula, response, n=100):
    ret = []
    
    for i in range(n):
        shuffled = data.sample(len(data))
        
        training, testing = shuffled.iloc[:len(gdata)//2], shuffled.iloc[len(gdata)//2:]
        
        trained = smf.ols(f_surviving, data=training).fit()
        resid = trained.predict(testing) - testing.babies_per_woman
        df = len(testing) - trained.df_model - 1
        mse = np.sum(resid**2)/ df
        
        ret.append(mse)
        
    return np.mean(ret)
```

```python
cross_validate(gdata, f_groupmeans, 'babies_per_woman')
```

```python
cross_validate(gdata, f_surviving, 'babies_per_woman')
```

```python
cross_validate(gdata, f_srp, 'babies_per_woman')
```
