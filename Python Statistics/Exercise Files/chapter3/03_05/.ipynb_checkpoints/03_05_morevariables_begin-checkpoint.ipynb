```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year]
    data.plot.scatter('gdp_per_day', 'life_expectancy', logx=True)
    
plotyear(1965)
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=False)
    area = 5e-6 * data.population
    color = data.age5_surviving
    data.plot.scatter('gdp_per_day', 'life_expectancy', logx=True, s=area,
                      c=color, colormap=matplotlib.cm.get_cmap('Purples_r'), 
                      vmin=55, vmax=100, linewidths=1, edgecolor='k', sharex=False)
    
plotyear(1965)
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=False)
    area = 5e-6 * data.population
    color = data.age5_surviving
    edgecolor = data.region.map({'Africa': 'skyblue', 'Europe': 'gold', 
                                'America': 'palegreen', 'Asia':'coral'})
    
    data.plot.scatter('gdp_per_day', 'life_expectancy', logx=True, s=area,
                      c=color, colormap=matplotlib.cm.get_cmap('Purples_r'), 
                      vmin=55, vmax=100, linewidths=1, edgecolors=edgecolor, 
                      sharex=False, figsize=(10, 6.5))
    
plotyear(1965)
```

```python
interact(plotyear,year=range(1965, 2016, 10))
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=False)
    area = 5e-6 * data.population
    color = data.age5_surviving
    edgecolor = data.region.map({'Africa': 'skyblue', 'Europe': 'gold', 
                                'America': 'palegreen', 'Asia':'coral'})
    
    data.plot.scatter('gdp_per_day', 'life_expectancy', logx=True, s=area,
                      c=color, colormap=matplotlib.cm.get_cmap('Purples_r'), 
                      vmin=55, vmax=100, linewidths=1, edgecolors=edgecolor, 
                      sharex=False, figsize=(10, 6.5))
    
    for level in [4,16,64]:
        pp.axvline(level, linestyle= ':', color='k')
    pp.axis(xmin=1, xmax=500, ymin=30, ymax=100)
    
plotyear(2015)
```

```python
gapminder['log10_gpd_per_day'] = np.log10(data['gdp_per_day'])
data = gapminder.loc[gapminder.year == 2015, ['log10_gpd_per_day', 'life_expectancy', 'age5_surviving',
                                             'babies_per_woman']]
pd.plotting.scatter_matrix(data, figsize=(9,9))
```
