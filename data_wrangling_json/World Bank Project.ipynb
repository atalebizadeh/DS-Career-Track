#### The counting method below will make sure that multiple projects from one code are only counted once.

```python
# code_nums is the highest number of mjtheme codes - 1 - 11
code_nums = 11

# dic will store the number of projects that refer to each code number.

dic = { }

# Initialize each code's count at zero. 
for i in range(1, code_nums+1):
    dic[i] = 0
    
    
# Move through each set of stored codes for each project to count the number of projects that include each code. 
for x in range(len(df)):
    # c stores the codes used for each project
    c = {}
    for y in range(len(df.mjtheme_namecode[x])):
        c[int(df.mjtheme_namecode[x][y]['code'])] = 1
        #print(c.keys())
    
    # Adds the c to overall count in md
    for i in c.keys():
        dic[i] += 1
    #print(dic)

#mj_df, stores the number of projects that refer to each code number, converting dic to a series 
# that I can now use Pandas to analyze the top occurences.
 
mj_df = pd.Series(dic)
mj_df.sort_values(ascending=False).head(10)
```
