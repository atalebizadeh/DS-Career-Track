### 1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).

```python
re = requests.get("https://www.quandl.com/api/v3/datasets/FSE/AFX_X?start_date=2017-01-01&end_date=2017-12-31&API_KEY")
re.json()
```

### 2. Convert the returned JSON object into a Python dictionary.

```python
dic = re.json()['dataset']
type(dic)
```

### 3. Calculate what the highest and lowest opening prices were for the stock in this period.

```python
# To get the opening price, we need to look at the second element under each datapoint (day)
dic['data'][1][1]
```

```python
# Using a for loop to go through the second element of each day and adding it to
# open_prices list
'''open_prices = []
for day in dic['data']:
    if day[1] != None:
        open_prices.append(day[1])
print(open_prices)'''

# Using list comprehensions to add second elemetn of each datapoint to
# a list
open_prices = [day[1] for day in dic['data'] if day[1] != None]
print(open_prices)
```

```python
max_open = str(max(open_prices))
min_open = str(min(open_prices))
print('The Highest opening price was: $' + max_open)
print('The lowest opening price was: $' + min_open )
```

### 4. What was the largest change in any one day (based on High and Low price)?

```python
# Accessing the high and low for a datapoint using the third and fourth elements
print('Datapoint High: '+ str(dic['data'][1][2]))
print('Datapoint low: '+ str(dic['data'][1][3]))
```

```python
high = dic['data'][1][2]
low = dic['data'][1][3]
day_change = high - low
print('Datapoint Day Change: '+ str(round(day_change,4)))
```

```python
# To get the daily change, we need to subtract the third element from the fourth one in our datapoints
# Then we use the round() and max() methods to get the largest change in any one day
daily_change = [day[2]-day[3] for day in dic['data'] if day[2:3] != None]
largest_change = str(round(max(daily_change), 2))
print('The largest change in any one day was: $' + largest_change)
```

### 5. What was the largest change between any two days (based on Closing Price)?

```python
# Using list comprehensions to add fifth elemetn of each datapoint to a list
close_prices = [day[4] for day in dic['data'] if day[4] != None]
```

```python
# Using for loop and enumerate in order to subtract an item (close price) with another item from the index before it. 
for index,item in enumerate(close_prices):
    print(index, item)
```

```python
for index,item in enumerate(close_prices):
    print(item - close_prices[index-1])
```

```python
'''two_days_change = []
for index, item in enumerate(close_prices):
    two_days_change.append(item - close_prices[index-1])'''
# We need to slice the list [1:] because if we don't, the list will subtract the first item (from end of 2017) from the 
# last item (beginning of 2017) in the list and return 15.96. Our data is sorted from end of 2017 to beginning of the year.
two_days_change = [item - close_prices[index-1] for index,item in enumerate(close_prices)][1:]
print(two_days_change)
```

```python
largest_change = str(round(max(two_days_change), 2))

print('The largest change between any two days was: $' + str(largest_change))
```

### 6.What was the average daily trading volume during this year?

```python
'''trade_volume = []
for day in dic['data']:
    if day[6]!= None:
        trade_volume.append(day[6])'''
# Using list comprehensions, we need to access the 7th element in our datapoints which represent the trade volume in each 
# day and add them to a list. Then using sum()/len() methods, we can get the average trade volume.
trade_volume = [day[6] for day in dic['data'] if day[6] != None]
trade_avg = round(sum(trade_volume)/len(trade_volume))
#print(trade_volume)
print('The average daily trading volume this year was: '+ str(trade_avg))
```
